```python
def no_norm(value,name,ampm):
    mn = pd.read_csv(f'mean_{ampm}.csv',index_col=0)
    mn = mn.T[name]
    std = pd.read_csv(f'std_{ampm}.csv',index_col=0)
    std = std.T[name]
    return (value*std).values[0]

def norm(value,name,ampm):
    mn = pd.read_csv(f'mean_{ampm}.csv',index_col=0)
    mn = mn.T[name].values
    std = pd.read_csv(f'std_{ampm}.csv',index_col=0)
    std = std.T[name].values
    return (value-mn)/std
```
